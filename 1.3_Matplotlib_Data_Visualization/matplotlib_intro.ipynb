````python
# 在实际使用的过程中，通常将matplotlib.pyplot缩写为plt。
import numpy as np
import matplotlib.pyplot as plt
````

````python
# 线图
# 基本格式 plt.plot(x, y, format_string, **kwargs)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

plt.plot(
    x,  # x坐标，默认值为range(len(y)),必须提供
    y,  # y坐标，必须提供
    "o-",  # 可选参数，定义图形样式
    label="Line Plot",  # 图例标签
    linewidth=2,  # 线宽
    color="blue",  # 线条颜色
    marker="s",  # 标记样式
    markersize=6,  # 标记大小
    markeredgecolor="red",  # 标记边缘颜色
    markeredgewidth=1,  # 标记边缘宽度
    markerfacecolor="yellow",  # 标记填充颜色
    alpha=1,  # 透明度
)
````

# 🎨 Matplotlib 常见颜色映射（Colormap）总表

---

## 🌈 1. 连续型（Sequential）
> 用于 **从小到大**、单调递增的数值（如温度、强度、浓度等）

| 名称 | 渐变方向 | 特点 | 说明 |
|------|-----------|------|------|
| `viridis` | 深蓝 → 绿 → 黄 | ✅ 默认推荐，视觉均匀、色盲友好 | Matplotlib 默认 cmap |
| `plasma` | 紫 → 橙黄 | 明亮、对比强 | 高可视度 |
| `inferno` | 深紫 → 红 → 黄 | 高对比、暗背景清晰 | 适合热力图 |
| `magma` | 黑紫 → 红橙 | 柔和渐变 | 暗色主题常用 |
| `cividis` | 蓝 → 黄 | 色盲友好（色觉无障碍优化） | 适合科学数据展示 |
| `Greys` | 黑 → 白 | 单色灰度 | 打印友好 |
| `Blues` | 浅蓝 → 深蓝 | 蓝色调渐变 | 气候、海洋数据 |
| `Oranges` | 浅橙 → 深橙 | 暖色调渐变 | 热能相关数据 |
| `Purples` | 浅紫 → 深紫 | 冷色系渐变 | 适合分类对比 |

---

## 🔀 2. 发散型（Diverging）
> 用于 **有中心值（如0）**，两侧正负对称变化的数据（如偏差、残差等）

| 名称 | 左侧 → 中心 → 右侧 | 特点 | 说明 |
|------|-------------------|------|------|
| `coolwarm` | 蓝 → 白 → 红 | 最常用发散型配色 | 平衡冷暖变化 |
| `bwr` | 蓝 → 白 → 红 | 与 `coolwarm` 类似，但色彩更鲜艳 | 简洁明快 |
| `seismic` | 蓝 → 白 → 红 | 对比强烈 | 地震、应力数据常用 |
| `RdBu` | 红 → 白 → 蓝 | 红蓝反向版本 | 政治、对立主题常见 |
| `PiYG` | 粉 → 白 → 绿 | 柔和渐变 | 适合对称指标对比 |

---

## 🧩 3. 离散/分类型（Qualitative）
> 用于 **类别数据**（如城市、物种、群体等）

| 名称 | 色调风格 | 特点 | 说明 |
|------|-----------|------|------|
| `tab10` | 10种明显区分的颜色 | Matplotlib 默认分类色 | 适合分类图 |
| `Set1` | 鲜艳红蓝绿等 | 高饱和度 | 适合少量类别 |
| `Set2` | 柔和色系 | 对比舒适 | 分类较多时使用 |
| `Accent` | 混合亮色 | 对比强烈 | 适合标注或突出数据 |
| `Dark2` | 深色调 | 稳重不刺眼 | 适合正式图表 |
| `Paired` | 成对对比 | 类别成双的情况 | 如“男/女”、“前/后” |

---

## 🗺️ 4. 特殊用途型（Miscellaneous）
> 一些早期或特定领域的映射，使用时需谨慎

| 名称 | 特点 | 说明 |
|------|------|------|
| `jet` | 彩虹色，视觉不均匀 ❌ | 不推荐，可能造成误读 |
| `rainbow` | 彩虹渐变 | 视觉艳丽但非线性 |
| `hot` | 黑 → 红 → 黄 → 白 | 常用于热度或能量数据 |
| `cool` | 青 → 品红 | 柔和过渡，适合渐变背景 |
| `terrain` | 绿色 → 棕 → 白 | 常用于地形/海拔图 |
| `ocean` | 蓝系 | 水文/气候地图 |
| `cubehelix` | 感知均匀灰度线性渐变 | 科学绘图常用 |



````python
# 计算正选曲线上的点的x和y的坐标
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# 创建一个颜色数组，基于y值映射到颜色
colors = y
plt.subplot(2, 1, 1)
# 绘制散点图1
plt.scatter(
    x,
    y,
    marker="o",
    c=colors,
    norm=None,
    vmin=-1,
    vmax=1,
    cmap="viridis",  # 使用'viridis'颜色映射
    alpha=0.8,
    linewidths=1,
    edgecolors="black",
)
plt.colorbar()  # 显示颜色条
plt.title("Scatter Plot of Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.subplot(2, 1, 2)
# 绘制散点图2
plt.scatter(
    x,
    y,
    s=50,  # 点的大小
    c=colors,  # 点的颜色
    marker="^",  # 点的形状
    alpha=0.6,  # 透明度
    edgecolors="black",  # 点的边缘颜色
    cmap="plasma",  # 使用'plasma'颜色映射
)
plt.colorbar()  # 显示颜色条
plt.title("Scatter Plot of Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# 调用 tight_layout 来自动调整子图参数
plt.tight_layout()
# 显示图形
plt.show()
````

````python
# 数据
sizes = [25, 35, 20, 20]
labels = ["A", "B", "C", "D"]
colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
# 创建饼图，并添加关键字参数
plt.pie(
    sizes,  # 饼图中每个扇形的尺寸
    explode=[0, 0, 0, 0],  # 用于指定每个扇形是否突出显示
    labels=labels,  # 用于指定每个扇形的标签
    colors=colors,  # 用于指定每个扇形的颜色
    autopct="%1.1f%%",  # 用于在饼图上显示每个扇形的百分比
    startangle=140,  # 饼图开始的角度，默认为 0（即从 x 轴正方向开始）
    shadow=False,  # 用于指定是否为饼图添加阴影
    radius=1,  # 饼图的半径
    wedgeprops=dict(
        edgecolor="black", linewidth=2, linestyle="-"
    ),  # 指定饼图中每个扇形的属性
    textprops=dict(
        color="red", weight="bold"
    ),  # 用于指定饼图中标签的文本属性，这里指定文本的颜色和字体的粗细
    center=(0, 0),  # 用于指定饼图的中心位置
    frame=False,  # 用于指定是否为饼图添加一个框
)
# 显示图表
plt.show()
````

## 🧩 什么时候才会新建画布？

在 Matplotlib 中，新画布（`Figure`）只有在以下两种情况下会被创建 👇

---

### ✅ 情况 1：手动创建新画布

使用 `plt.figure()` 会**显式创建一个新的画布**。

每次调用 `plt.figure()` 都会生成一个**独立的图窗口**。

```python
import matplotlib.pyplot as plt

plt.figure()       # 手动新建一个 Figure
plt.plot([1, 2, 3])
plt.show()
```
## ✅ 情况 2：前一个 Figure 已经显示并关闭

如果你调用了 `plt.show()`  **并关闭了图窗口**，
下一次绘图时，`Matplotlib` **会自动创建一个新的** `Figure`。
```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3])
plt.show()    # 显示并关闭第一个 Figure

plt.plot([4, 5, 6])
plt.show()    # 自动创建第二个 Figure

```
🔍 因此，看起来好像“每次 show() 都新建了画布”，
但实际上是：

- show() 不会创建新画布

- 而是因为上一个画布被关闭后，下一次绘图自动生成新的 Figure
---

| 情况                 | 是否创建新画布 | 说明            |
| ------------------ | ------- | ------------- |
| 第一次调用 `plt.plot()` | ✅ 是     | 自动创建默认 Figure |
| 再次绘图（未关闭）          | ❌ 否     | 使用当前 Figure   |
| 调用 `plt.figure()`  | ✅ 是     | 手动新建 Figure   |
| `plt.show()`       | ❌ 否     | 仅显示，不创建       |
| 关闭图后再次绘图           | ✅ 是     | 自动新建新 Figure  |


````python
# 创建一个Figure对象
plt.figure(
    # num=None,
    # figsize=None,
    # dpi=None,
    # facecolor=None,
    # edgecolor=None,
    # frameon=True,
    # clear=False,
    # **kwargs
)  # 设置图形大小为10x6英寸
````

⚙️ 参数说明（含默认值）

| 参数            | 类型                | 默认值                     | 说明                                                                 |
| ------------- | ----------------- | ----------------------- | ------------------------------------------------------------------ |
| **num**       | 整数 / 字符串 / `None` | `None`                  | 图形窗口的编号或名称。<br> - 若为 `None` → 创建新窗口。<br> - 若编号/名称已存在 → 获取并激活该图形窗口。 |
| **figsize**   | 元组 `(宽, 高)`（英寸）   | `None`（默认：`(6.4, 4.8)`） | 指定图形窗口大小。示例：`figsize=(8, 6)`                                       |
| **dpi**       | 数值（每英寸点数）         | `None`（默认：`100`）        | 设置图形分辨率。<br>数值越高，图像越清晰，文件越大。                                       |
| **facecolor** | 颜色字符串 / RGB值      | `None`（默认：白色）           | 设置画布（背景）的填充颜色。<br>如 `'lightgray'`、`'#f0f0f0'`、`(0.9, 0.9, 0.9)`    |
| **edgecolor** | 颜色字符串 / RGB值      | `None`                  | 设置画布边缘的颜色。                                                         |
| **frameon**   | 布尔值               | `True`                  | 是否绘制图形窗口边框。`False` 表示不显示外边框。                                       |
| **clear**     | 布尔值               | `False`                 | 若图形窗口已存在：<br> - `True` → 清除原有内容；<br> - `False` → 保留现有图像。           |
| **kwargs**    | 其他关键字参数           | —                       | 其他可选参数，用于进一步自定义 Figure 属性。                                         |

🎨 参数效果示意

| 参数                        | 作用     | 示例效果      |
| ------------------------- | ------ | --------- |
| `figsize=(10,6)`          | 控制画布大小 | 图像更宽或更高   |
| `dpi=200`                 | 控制分辨率  | 导出的图片更清晰  |
| `facecolor='lightyellow'` | 设置背景色  | 改变绘图区背景   |
| `edgecolor='black'`       | 设置边框颜色 | 仅在某些后端可见  |
| `clear=True`              | 清空已有图像 | 重绘新内容而非叠加 |

🧩 进阶技巧

| 功能      | 示例                             | 说明                         |
| ------- | ------------------------------ | -------------------------- |
| 获取当前图对象 | `fig = plt.gcf()`              | `gcf` = get current figure |
| 创建多个图   | `plt.figure(1); plt.figure(2)` | 可同时管理多个画布                  |
| 清空当前图   | `plt.clf()`                    | 清除当前画布所有内容                 |
| 关闭图形    | `plt.close()`                  | 关闭窗口（可指定编号）                |


🧾 总结

| 类别   | 参数示例                                | 默认值                | 功能         |
| ---- | ----------------------------------- | ------------------ | ---------- |
| 图像识别 | `num`                               | `None`             | 指定图窗口编号或名称 |
| 尺寸控制 | `figsize`, `dpi`                    | `(6.4,4.8)`, `100` | 控制大小与清晰度   |
| 样式外观 | `facecolor`, `edgecolor`, `frameon` | `白色`, `True`       | 控制背景与边框    |

# 🧩 Matplotlib — `plt.subplots()` 函数详解（含默认值）

---

## 🖼️ 函数原型

```python
plt.subplots(
    nrows=1,
    ncols=1,
    sharex=False,
    sharey=False,
    squeeze=True,
    subplot_kw=None,
    gridspec_kw=None,
    **fig_kw
)
```
## ⚙️ 参数说明（含默认值）

| 参数                | 类型         | 默认值     | 说明                                                                              |
| ----------------- | ---------- | ------- | ------------------------------------------------------------------------------- |
| **nrows**         | 整数         | `1`     | 子图的行数。设置为 1 表示只有一行。                                                             |
| **ncols**         | 整数         | `1`     | 子图的列数。设置为 1 表示只有一列。                                                             |
| **sharex**        | 布尔值 / 字符串  | `False` | 是否共享 x 轴：<br> - `True`：所有子图共享相同的 x 轴刻度。<br> - `'col'`：按列共享。<br> - `'row'`：按行共享。 |
| **sharey**        | 布尔值 / 字符串  | `False` | 是否共享 y 轴：<br> - `True`：所有子图共享相同的 y 轴刻度。<br> - `'col'` / `'row'` 同理。             |
| **squeeze**       | 布尔值        | `True`  | 是否压缩返回的数组：<br> - 若只创建一个子图，返回单个 `Axes` 对象。<br> - 若创建多图，返回二维数组。                   |
| **width_ratios**  | 序列或 `None` | `None`  | 控制每一列的宽度比例（例如 `[2, 1]`）。                                                        |
| **height_ratios** | 序列或 `None` | `None`  | 控制每一行的高度比例（例如 `[1, 3]`）。                                                        |
| **subplot_kw**    | 字典         | `None`  | 传递给每个 `add_subplot()` 的额外参数。<br>例如：`subplot_kw={'projection': 'polar'}` 用于极坐标。  |
| **gridspec_kw**   | 字典         | `None`  | 控制底层网格（GridSpec）的布局属性。<br>例如：`{'wspace': 0.3, 'hspace': 0.4}` 调整间距。             |
| **fig_kw**        | 关键字参数      | —       | 传递给 `plt.figure()` 的参数，如：`figsize`、`dpi`、`facecolor` 等。                         |


## 🧠 返回值

fig, axes = plt.subplots(...)

| 返回对象     | 类型                                 | 说明                               |
| -------- | ---------------------------------- | -------------------------------- |
| **fig**  | `matplotlib.figure.Figure`         | 整个画布对象                           |
| **axes** | `matplotlib.axes.Axes` 或 `ndarray` | 一个或多个子图对象（取决于 `nrows` × `ncols`） |


````python
# 创建数据
x = np.arange(0, 10, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = 1 / (1 + np.exp(-x))  # Sigmoid function
# 创建一个 2x2 的子图网格
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
# 第一个子图
axs[0, 0].plot(x, y1)
axs[0, 0].set_title("Sine Wave")
axs[0, 0].set_xlabel("X-axis")
axs[0, 0].set_ylabel("Y-axis")
# 第二个子图
axs[0, 1].plot(x, y2)
axs[0, 1].set_title("Cosine Wave")
axs[0, 1].set_xlabel("X-axis")
axs[0, 1].set_ylabel("Y-axis")
# 第三个子图
axs[1, 0].plot(x, y3)
axs[1, 0].set_title("Tangent Wave")
axs[1, 0].set_xlabel("X-axis")
axs[1, 0].set_ylabel("Y-axis")
# 第四个子图
axs[1, 1].plot(x, y4)
axs[1, 1].set_title("Sigmoid Function")
axs[1, 1].set_xlabel("X-axis")
axs[1, 1].set_ylabel("Y-axis")
# 调整子图间距
plt.tight_layout()
# 显示图形
plt.show()
````

```python
fig.add_subplot(nrows, ncols, plot_number)
```
- nrows：子图的总行数。
- ncols：子图的总列数。
- plot_number：当前子图在其所在布局中的位置，从1开始计数，从左上角到右下角。函数会返回一个Axes对象

````python
# 创建一个图形窗口
fig = plt.figure()
# 添加第一个子图，位于2行1列布局中的第1个位置
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot([0, 1], [0, 1])
# 添加第二个子图，位于2行1列布局中的第2个位置
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot([0, 1], [1, 0])
# 显示图形
plt.show()
````

## add_axes
```python
matplotlib.figure.Figure.add_axes(rect, **kwargs)
```
- rect：一个形如 [左边界, 底边界, 宽度, 高度] 的列表，指定子图的位置和大小。这些值是在归一化坐标中的，其中 0 表示图形的左边界，1 表示右边界，0表示底部，1 表示顶部。
- **kwargs：其他可选的关键字参数，用于控制子图的属性。

````python
# 创建一个图形
fig = plt.figure()
# 添加一个子图，位置是图形左下角的1/4处，大小为图形宽度的1/2和高度的1/2
ax = fig.add_axes([0.1, 0.1, 0.5, 0.5])
# 使用这个子图绘制一些数据
ax.plot([0, 1], [0, 1])
# 显示图形
plt.show()
````

###  suptitle
```python
matplotlib.figure.Figure.suptitle(t, **kwargs)
```
- t：字符串，表示要添加的标题文本。
- **kwargs：其他可选的关键字参数，用于控制标题的样式，如字体大小（ fontsize）、颜色（ color）等。

````python
# 创建一个图形
fig = plt.figure()
# 在图形上添加两个子图
ax1 = fig.add_subplot(211)  # 第一个子图，两行一列的第一个位置
ax2 = fig.add_subplot(212)  # 第二个子图，两行一列的第二个位置
# 使用第一个子图绘制一些数据
ax1.plot([0, 1], [0, 1])
# 使用第二个子图绘制一些数据
ax2.plot([0, 1], [1, 0])
# 为整个图形添加一个总标题
fig.suptitle("Test Suptitle", fontsize=16, color="red")
# 显示图形
plt.show()
````

## text
```python
matplotlib.axes.Axes.text(x, y, s, **kwargs)

```
- x, y：浮点数或数组，指定文本的位置坐标。
- s：字符串，要添加的文本内容。
- **kwargs：其他可选的关键字参数，用于控制文本的样式，如 fontsize（字体大小）、color（颜色）、 horizontalalignment（水平对齐方式，如 ‘left’,‘center’, ‘right’）、 verticalalignment（垂直对齐式，如 ‘top’, ‘center’,‘bottom’）等。

````python
# 创建一个图形和一个子图
fig, ax = plt.subplots()
# 绘制一些数据
ax.plot([0, 1, 2], [0, 1, 0])
# 在坐标 (1, 0.5) 处添加文本
ax.text(
    1,
    0.5,
    "Sample Text",
    fontsize=12,
    color="red",
    horizontalalignment="center",
    verticalalignment="center",
)
# 显示图形
plt.show()
````

````python
# 创建一个图形和一个子图
fig, ax = plt.subplots(2, 2)
# 获取图形中的所有轴
axes = fig.axes
# axes 是一个包含所有轴的列表
for ax in axes:
    # 包含了每一个轴的x和y的相对位置(从左下角开始计算)、轴的宽度和高度(相对于整个图形)
    print(ax)
# 显示图像
plt.show()
````

````python
# get_facecolor 该函数是一个用于获取轴（Axes）或填充区域（Patch）背景色的方法。
# 创建一个图形
fig = plt.figure(figsize=(8, 6), facecolor="green")
# 获取背景色
facecolor = fig.get_facecolor()
print(facecolor)
plt.plot([1, 2, 3], [4, 5, 6])
# 显示图像
plt.show()
````

````python
# get_dpi 该方法用于获取图形（Figure）对象的分辨率。
# 创建一个图形对象
fig = plt.figure()
# 获取图形的DPI
dpi = fig.get_dpi()
print(f"The DPI of the figure is: {dpi}")
# 显示图像
plt.show()
````

````python
# tight_layout 该函数用于自动调整子图参数，使之填充整个图像区域，同时确保子图之间的标签和标题不会重叠。这个方法对于创建布局整齐的图形非常有用，尤其是在子图较多或标签较长时

# 创建一个包含多个子图的图形
fig, axs = plt.subplots(2, 2)
# 在每个子图中绘制一些数据
axs[0, 0].plot([1, 2, 3], [1, 2, 3])
axs[0, 1].plot([1, 2, 3], [3, 2, 1])
axs[1, 0].plot([1, 2, 3], [1, 3, 2])
axs[1, 1].plot([1, 2, 3], [2, 1, 3])
# 使用 tight_layout 自动调整子图布局
fig.tight_layout()
# 显示图形
plt.show()
````

### subplots_adjust  该函数允许用户手动调整子图之间的空间，以及子图与图形边缘之间的空间。这个方法提供了比 tight_layout 更细粒度的控制，特别适用于需要精确控制布局的情况
```python
fig.subplots_adjust(left=None, bottom=None, right=None, top=None,wspace=None, hspace=None)
```
- left : 子图区域左边缘与图形左边缘之间的距离，范围从 0 到 1。
- bottom : 子图区域下边缘与图形下边缘之间的距离，范围从 0 到 1。
- right : 子图区域右边缘与图形右边缘之间的距离，范围从 0 到 1。
- top : 子图区域上边缘与图形上边缘之间的距离，范围从 0 到 1。
- wspace : 子图之间的水平间距，范围从 0 到 1。
- hspace : 子图之间的垂直间距，范围从 0 到 1。

![image.png](../image/subplot_adjust.png)

````python
# 创建一个包含多个子图的图形
fig, axs = plt.subplots(2, 2)
# 在每个子图中绘制一些数据
axs[0, 0].plot([1, 2, 3], [1, 2, 3])
axs[0, 1].plot([1, 2, 3], [3, 2, 1])
axs[1, 0].plot([1, 2, 3], [1, 3, 2])
axs[1, 1].plot([1, 2, 3], [2, 1, 3])
# 使用 subplots_adjust 手动调整子图布局
fig.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, hspace=0.5, wspace=0.5)
# 显示图形
plt.show()
````

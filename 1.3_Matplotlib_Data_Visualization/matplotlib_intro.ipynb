## ðŸ—ºï¸ 4. ç‰¹æ®Šç”¨é€”åž‹ï¼ˆMiscellaneousï¼‰
> ä¸€äº›æ—©æœŸæˆ–ç‰¹å®šé¢†åŸŸçš„æ˜ å°„ï¼Œä½¿ç”¨æ—¶éœ€è°¨æ…Ž

### subplots_adjust  è¯¥å‡½æ•°å…è®¸ç”¨æˆ·æ‰‹åŠ¨è°ƒæ•´å­å›¾ä¹‹é—´çš„ç©ºé—´ï¼Œä»¥åŠå­å›¾ä¸Žå›¾å½¢è¾¹ç¼˜ä¹‹é—´çš„ç©ºé—´ã€‚è¿™ä¸ªæ–¹æ³•æä¾›äº†æ¯” tight_layout æ›´ç»†ç²’åº¦çš„æŽ§åˆ¶ï¼Œç‰¹åˆ«é€‚ç”¨äºŽéœ€è¦ç²¾ç¡®æŽ§åˆ¶å¸ƒå±€çš„æƒ…å†µ

- left : å­å›¾åŒºåŸŸå·¦è¾¹ç¼˜ä¸Žå›¾å½¢å·¦è¾¹ç¼˜ä¹‹é—´çš„è·ç¦»ï¼ŒèŒƒå›´ä»Ž 0 åˆ° 1ã€‚

- right : å­å›¾åŒºåŸŸå³è¾¹ç¼˜ä¸Žå›¾å½¢å³è¾¹ç¼˜ä¹‹é—´çš„è·ç¦»ï¼ŒèŒƒå›´ä»Ž 0 åˆ° 1ã€‚
- top : å­å›¾åŒºåŸŸä¸Šè¾¹ç¼˜ä¸Žå›¾å½¢ä¸Šè¾¹ç¼˜ä¹‹é—´çš„è·ç¦»ï¼ŒèŒƒå›´ä»Ž 0 åˆ° 1ã€‚
